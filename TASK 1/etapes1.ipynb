```python
import pandas as pd
```

```python
df = pd.read_excel("KPMG_VI_New_raw_data_update_final.xlsx", sheet_name="CustomerDemographic", header=1)
df
```

# Vérifier les valeurs uniques d'une colonne

```python
print(df['gender'].unique())
print(df['deceased_indicator'].unique())
print(df['owns_car'].unique())
```

```python
# Compter les valeurs manquantes par colonne
print(df.isnull().sum())
```

```python
# Essayez de convertir la colonne DOB en datetime
try:
    df['DOB'] = pd.to_datetime(df['DOB'])
    print("Toutes les dates sont valides.")
except ValueError as e:
    print(f"Erreurs de dates détectées : {e}")
```

```python
# Vérification de l'âge
min_age = 0  # L'âge minimal attendu
max_age = 100  # L'âge maximal attendu (ou toute autre limite raisonnable)

# Vérifiez si les âges sont valides
invalid_ages = df[(df['DOB'].dt.year < (pd.Timestamp.now().year - max_age)) | (df['DOB'].dt.year > (pd.Timestamp.now().year - min_age))]
if not invalid_ages.empty:
    print("Erreurs d'âge détectées :")
    print(invalid_ages[['customer_id', 'DOB']])
else:
    print("Tous les âges sont conformes à la logique métier.")
```
